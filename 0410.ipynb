```python
from skimage import io,data

img = io.imread('QQ.jpg')
rows,cols,dims=img.shape

img2 = img[:,:,0]
for i in range(0,rows):
    for j in range(0,cols):
        sum = 0.0
        for k in range(0,3): #dims = 3
            sum = sum + img[i,j,k]
        img2[i,j] = sum/3 #dims = 3
io.imshow(img2)
io.show()

io.imsave('QQ3.jpg',img2)
```

```python
from skimage import io,data
import matplotlib.pyplot as plt
import numpy as np

img = io.imread('QQ.jpg')
rows,cols,dims=img.shape

img2 = np.uint8(np.zeros((rows,cols,3)))

plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(img)

for i in range(0,rows):
    for j in range(0,cols):
        sum = 0.0
        for k in range(0,3):
            sum = sum + img[i,j,k]
        for k in range(0,3):
            img2[i,j,k] = np.uint8(sum/3)
        
plt.subplot(1,2,2)
plt.title('grey image')
plt.imshow(img2)
plt.show()
```
